# How to Read, Write, Show Videos from Camera in OpenCV

```python
cap = cv2.VideoCapture(0) # 0 means the default camera
fourcc = cv2.VideoWriter_fourcc(*'XVID') # fourcc is a 4-byte code used to specify the video codec
out = cv2.VideoWriter('output.avi' , fourcc , 20.0 , (640 , 480)) # 20.0 is the frame rate , (640 , 480) is the frame size

print(cap.isOpened()) # check if the camera is opened or not

while(cap.isOpened()):
    ret , frame = cap.read() # ret is a boolean variable which returns true if the frame is available
    if ret == True : 
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # get the width of the frame
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) # get the height of the frame
        
        out.write(frame) # write the frame into the output file
        
        gray = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY) # convert the frame into gray scale
        
        cv2.imshow('frame' , gray) # display the frame
        
        if cv2.waitKey(1) == ord('q'):
            break
    else:
        break
    

cap.release() # release the camera
out.release() # release the output file
cv2.destroyAllWindows() # close all the windows
```

# Create Graphics with opencv2 


```python
import numpy as np

import cv2 

# create a image with the (600,300)
img = np.zeros((600,900,3) , np.uint8)

cv2.rectangle(img , (0,0) , (900,400) , (255,255,85) , -1)
cv2.rectangle(img , (0,400) , (900,600) , (75,180,70) , -1)

# DRAW SUN
cv2.circle(img , (100,100) , 50 , (0,255,255) , -1)
# draw tree 
cv2.rectangle(img , (400,200) , (450,400) , (0,255,0) , -1)


triangle2 = np.array([[640,460],[780,460], [710,200]], dtype=np.int32) #array to darw fillpoly

cv2.fillPoly(img, [triangle2], (75,180,70)) # fillpoly to draw triangle

cv2.imshow('myimage' , img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Setting Camera Parameters in OpenCV Python

```python
import cv2
cap = cv2.VideoCapture(0)

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 3000)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 3000)


while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:

       gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
       cv2.imshow('frame', gray)

       if cv2.waitKey(1) & 0xFF == ord('q'):
         break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

# Show Date and Time on Videos using OpenCV Python

```python
import cv2 
import datetime 

cap = cv2.VideoCapture(0)
cap.set(3 , 3000)
while(cap.isOpened()):
    ret , frame = cap.read()
    
    if ret == True : 
        
        font = cv2.FONT_HERSHEY_SIMPLEX
        text = 'Width : ' +str(cap.get(3)) +'Hieght:' +str(cap.get(4))
        
        datet = str(datetime.datetime.now())
        # Add text 
        
        frame = cv2.putText( frame , text , (10,50) , font , 1 , (0,255,255) , 1)
        frame = cv2.putText( frame , datet , (10,100) , font , 1 , (0,255,255) , 1)
        
        cv2.imshow('frame' , frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'): 
            break
    else :
        break

cap.release()
cv2.destroyAllWindows()
        
        
```
